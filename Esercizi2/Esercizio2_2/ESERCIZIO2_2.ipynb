```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file_path = "/home/simonelance/Documents/MAGISTRALE/LABO DI GALLI/Esercizi/data/risultati2_2_1.out"


df = pd.read_csv(file_path, sep=' ' )

# Mostra le prime righe per verificare
print(df.head())
```

```python
plt.figure(figsize=(8, 6))

# Plotta con barre d'errore
plt.errorbar(df['n_step'], df['r'], yerr=df['error'], capsize=1, label='Dati con errore')# fmt='o',

# Personalizza il grafico
plt.xlabel('Numero di passi (n_step)')
plt.ylabel('Distanza (r)')
plt.title('Grafico di r vs n_step con barre di errore')
plt.legend()
plt.grid(True)

# Mostra il grafico
plt.show()
```
