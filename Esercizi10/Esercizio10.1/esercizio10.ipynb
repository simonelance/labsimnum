```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path= "OUTPUT/core0/path_0.out"
df = pd.read_csv(path ,delim_whitespace=True)
path500= "OUTPUT/core0/path_10000.out"
df500 = pd.read_csv(path500 ,delim_whitespace=True)

# Rinomina le colonne

print(df)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 8))  # 1 riga, 2 colonne

# --- Primo subplot: df ---
axes[0].scatter(df['X'].tail(34), df['Y'].tail(34), s=8, edgecolor='r')
axes[0].plot(df['X'].tail(34), df['Y'].tail(34), color='blue', linewidth=1)
idx1 = df.index[df['city_id'] == 1][0]
axes[0].scatter(df.loc[idx1, 'X'], df.loc[idx1, 'Y'],
                color='gold', edgecolor='black', s=100, marker='*', label='Primo punto')
axes[0].set_title("df - Distribuzione Finale delle città")
axes[0].legend()

# --- Secondo subplot: df500 ---
axes[1].scatter(df500['X'].tail(34), df500['Y'].tail(34), s=8, edgecolor='r')
axes[1].plot(df500['X'].tail(34), df500['Y'].tail(34), color='blue', linewidth=1)
idx2 = df500.index[df500['city_id'] == 1][0]
axes[1].scatter(df500.loc[idx2, 'X'], df500.loc[idx2, 'Y'],
                color='gold', edgecolor='black', s=100, marker='*', label='Primo punto')
axes[1].set_title("df500 - Distribuzione delle città al passo 250")
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
pt = df[df['city_id'] == 1]
plt.figure(figsize=(10,9))
plt.scatter(pt['X'], pt['Y'],
            color='gold', edgecolor='black',
            s=100, marker='*', label='Punto id=1')
plt.legend()


plt.scatter(df['X'].tail(34), df['Y'].tail(34), s = 8, edgecolor = 'r' )
plt.plot(df['X'].tail(34), df['Y'].tail(34), color='blue', linewidth=1)

plt.title("Distribuzione delle città sulla circonf unitaria")
plt.show()
```
