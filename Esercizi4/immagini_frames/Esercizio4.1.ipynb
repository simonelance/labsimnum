```python
import matplotlib.pyplot as plt


import pandas as pd
import numpy as np
```

```python
file_path = "NSL_SIMULATOR/OUTPUT/pofv.dat"
```

```python
df = pd.read_csv(file_path, sep='\s+' )
df
```

```python
x = df['#'].head(30)  # Numero di passi
y = df['VELOCITY:'].head(30)  # Valori della velocità
yerr = df['ERROR:'].head(30)  # Errori

plt.figure(figsize=(8, 6))

# Plotta l'istogramma
plt.bar(x, y, yerr=yerr, width=0.1	, capsize=3, color='lightblue', edgecolor='black', label='Dati con errore')


# Personalizza il grafico
plt.xlabel('Numero di passi (n_step)')
plt.ylabel('VELOCITY COUNTER')
plt.title('# DI PARTICELLE PER RANGE DI VELOCITÀ')
plt.legend()

plt.grid(True)

# Mostra il grafico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import imageio

# Supponiamo che `df` sia già caricato
batch_size = 30  # Numero di valori per ciascun istogramma
num_batches = len(df) // batch_size  # Numero totale di gruppi

# Lista per salvare i nomi dei frame temporanei
frames = []

for i in range(num_batches):
    start = i * batch_size
    end = start + batch_size
    x = df['#'][start:end]  # Estrazione del sottoinsieme di dati
    y = df['AVE_POFV:'][start:end]
    yerr = df['ERROR:'][start:end]

    plt.figure(figsize=(8, 6))
    plt.bar(x, y, yerr=yerr, width=0.1, capsize=3, color='lightblue', edgecolor='black', label=f'Batch {i+1}')
    
    plt.xlabel('Numero di passi (n_step)')
    plt.ylabel('VELOCITY COUNTER')
    plt.title(f'# DI PARTICELLE PER RANGE DI VELOCITÀ - Step {i+1}')
    plt.legend()
    plt.grid(True)
    
    filename = f'frame_{i}.png'
    plt.savefig(filename)
    frames.append(filename)
    plt.close()  # Chiude la figura per risparmiare memoria

# Creazione della GIF
gif_filename = "istogramma_animato.gif"
imageio.mimsave(gif_filename, [imageio.imread(frame) for frame in frames], duration=500, loop=0)

print(f"GIF salvata come {gif_filename}")
```
