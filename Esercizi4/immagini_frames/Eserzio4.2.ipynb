```python
import matplotlib.pyplot as plt
import os

import pandas as pd
import numpy as np
```

```python
file_path = "NSL_SIMULATOR/OUTPUT/FILES_4.2/pofv_transition.dat"
U_path = "NSL_SIMULATOR/OUTPUT/FILES_4.2/potential_energy.dat"
temperature_path ="NSL_SIMULATOR/OUTPUT/FILES_4.2/temperature.dat"
```

```python
df = pd.read_csv(file_path, sep='\s+' )
df.head(35)
```

```python
U =pd.read_csv("NSL_SIMULATOR/OUTPUT/FILES_4.2/potential_energy.dat", sep='\s+' )
x = U['BLOCK:']  # Numero di passi dalla riga 30 alla 60 (escluso 60)
y = U['ACTUAL_PE:']  # Velocit dalla riga 30 alla 60

plt.figure(figsize=(10, 6))

# Plotta l'istogramma
plt.plot(x, y, color='blue', label='Energia Potenziale',marker = 'o')  # Grafico a linee
#plt.axhline(y=2.19, color='red', linestyle='--', label='Linea a y = 2.19')


# Personalizza il grafico
plt.xlabel('N BLOCCHI')
plt.ylabel('En. Potenziale')
plt.title('CONTROLLO ENERGIA POTENZIALE PER VERIFICARE EQUILIBRIO')
plt.legend()

plt.grid(True)

# Mostra il grafico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

U = pd.read_csv(U_path, sep='\s+')

x1 = U['BLOCK:']
y1 = U['ACTUAL_PE:']

x2 = U['BLOCK:']
y2 = U['ACTUAL_PE:']

plt.figure(figsize=(8, 4))

# Disegna entrambe le serie sullo stesso grafico
plt.plot(x1, y1, color='green', marker='o', label='Blocchi 050 (inizio)')
plt.plot(x2, y2, color='red', marker='o', label='Blocchi 50100 (dopo equilibrio)')
#plt.axhline(y=2.19, color='blue', linestyle='--', label='y = 2.19')

plt.title('Confronto fluttuazioni di U - prima e dopo equilibrio')
plt.xlabel('# Blocchi')
plt.ylabel('U')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
temperature =pd.read_csv(temperature_path, sep='\s+' )
x = temperature['BLOCK:']  # Numero di passi dalla riga 30 alla 60 (escluso 60)
y = temperature['ACTUAL_T:']  # Velocit dalla riga 30 alla 60

plt.figure(figsize=(8, 6))

# Plotta l'istogramma
plt.plot(x, y, color='blue', label='Temperatura',marker = 'o')  # Grafico a linee


# Personalizza il grafico
plt.xlabel('# BLOCCHI')
plt.ylabel('TEMPERATURE')
plt.title('CONTROLLO TEMPERATURE PER VERIFICARE EQUILIBRIO')
plt.legend()

plt.grid(True)

# Mostra il grafico
plt.show()
```

```python
x1 = df['VELOCITY:'].iloc[0:30]
y1 = df['AVE_POFV:'].iloc[0:30]
yerr1 = df['ERROR:'].iloc[0:30]

x2 = df['VELOCITY:'].iloc[-30:]
y2 = df['AVE_POFV:'].iloc[-30:]
yerr2 = df['ERROR:'].iloc[-30:]

# Crea la figura con 2 subplot affiancati
fig, axs = plt.subplots(1, 2, figsize=(16, 6), sharey=True)  # sharey mantiene la stessa scala verticale

# Primo istogramma (a sinistra)
axs[0].bar(x1, y1, yerr=yerr1, width=0.1, capsize=3, color='lightblue', edgecolor='black')
axs[0].set_title('# DI PARTICELLE - Inizio')
axs[0].set_xlabel('Velocit (primi 30)')
axs[0].set_ylabel('VELOCITY COUNTER')
axs[0].grid(True)

# Secondo istogramma (a destra)
axs[1].bar(x2, y2, yerr=yerr2, width=0.1, capsize=3, color='salmon', edgecolor='black')
axs[1].set_title('# DI PARTICELLE - Fine')
axs[1].set_xlabel('Velocit (ultimi 30)')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
!pip install imageio

import imageio

# Supponiamo che `df` sia gi caricato
batch_size = 30  # Numero di valori per ciascun istogramma
num_batches = len(df) // batch_size  # Numero totale di gruppi

# Lista per salvare i nomi dei frame temporanei
frames = []
output_folder = 'immagini_frames'  # Puoi scegliere il nome della cartella
os.makedirs(output_folder, exist_ok=True)  # Crea la cartella se non esiste

x_min = df['VELOCITY:'].min()  # Puoi sostituirlo con un valore fisso se conosci il range
x_max = df['VELOCITY:'].max()  # Puoi sostituirlo con un valore fisso se conosci il range
y_min = df['AVE_POFV:'].min() - 0.05  # Un po' di margine sotto il minimo
y_max = df['AVE_POFV:'].max() + 0.01  # Un po' di margine sopra il massimo


for i in range(num_batches):
    start = i * batch_size
    end = start + batch_size
    x = df['VELOCITY:'][start:end]  # Estrazione del sottoinsieme di dati
    y = df['AVE_POFV:'][start:end]
    yerr = df['ERROR:'][start:end]

    plt.figure(figsize=(8, 6))
    plt.bar(x, y, yerr=yerr, width=0.1, capsize=3, color='lightblue', edgecolor='black', label=f'Batch {i+1}')
    
    plt.xlabel('Numero di passi (n_step)')
    plt.ylabel('VELOCITY COUNTER')
    plt.title(f'# DI PARTICELLE PER RANGE DI VELOCIT - Step {i+1}')
    plt.legend()
    plt.grid(True)
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)

    
    filename = os.path.join(output_folder, f'frame_{i}.png')    
    plt.savefig(filename)
    frames.append(filename)
    plt.close()  # Chiude la figura per risparmiare memoria

# Creazione della GIF
gif_filename = "istogramma_animato.gif"
imageio.mimsave(gif_filename, [imageio.imread(frame) for frame in frames], duration=500, loop=0)

print(f"GIF salvata come {gif_filename}")
```
